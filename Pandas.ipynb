# Conditional Changes

```python
# change 'Fire' to 'Flamer' in Type 1 column
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df
# revert changes
df.loc[df['Type 1'] == 'Flamer', 'Type 1'] = 'Fire'
df.head(5)

# change 'Fire' in 'Type 1' column should be 'true' in 'Legendary' column
df.loc[df['Type 1'] == 'Fire', 'Legendary'] = 'True'
df.head(5)
```
